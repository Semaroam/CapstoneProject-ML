```python
tourism_all = np.concatenate((
    info_tourism.Place_Id.unique(),
    tourism_rating.Place_Id.unique()
))

tourism_all = np.sort(np.unique(tourism_all))

print(f"Total number of tourism: {len(tourism_all)}")
```

```python
all_tourism_rate = tourism_rating
all_tourism_rate
```

```python
# merge the tourism information with the rating
all_tourism = pd.merge(all_tourism_rate,info_tourism[["Place_Id","Place_Name","Description","City","Category"]],on='Place_Id', how='left')
all_tourism
```

```python
# only use place_name, description, and city
all_tourism = all_tourism[['Place_Id','Place_Ratings','Place_Name','Description','City','Category']]
all_tourism
```

```python
# only take the semarang data from the crawl wisata
all_tourism_semarang = all_tourism[all_tourism.City == "Semarang"]
all_tourism_semarang
```

```python
print(f"Total duplicate data: {all_tourism_semarang.duplicated().sum()}")
```

```python
# drop the missing value
all_tourism_semarang.dropna(inplace=True)
```

```python
# drop the duplicate data
all_tourism_semarang.drop_duplicates(inplace=True)
print(f"Total duplicate data: {all_tourism_semarang.duplicated().sum()}")
```
